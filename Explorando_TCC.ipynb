```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv("/content/reading_habits.csv")
df
```

# **Traduzindo os nomes das colunas**

```python
traducao = {
    'Age': 'Idade',
    'Sex': 'Sexo',
    'Race': 'Raça',
    'Marital status?': 'Estado civil',
    'Education': 'Educação',
    'Employement': 'Emprego',
    'Incomes': 'Salario',
    'How many books did you read during last 12months?': 'Quantos livros você leu nos últimos 12 meses?',
    'Read any printed books during last 12months?': 'Leu algum livro físico nos últimos 12 meses?',
    'Read any audiobooks during last 12months?': 'Leu algum audiolivro nos últimos 12 meses?',
    'Read any e-books during last 12months?': 'Leu algum e-book nos últimos 12 meses?',
    'Last book you read, you…': 'No último livro que você leu, você…',
    'Do you happen to read any daily news or newspapers?': 'Você costuma ler notícias ou jornais diariamente?',
    'Do you happen to read any magazines or journals?': 'Você costuma ler revistas?'
}

df.rename(columns=traducao, inplace=True)
```

# **Traduzindo os valores**

```python
df.replace("No", "Não", inplace=True)
df.replace("Yes", "Sim", inplace=True)

# Sexo
df.replace("Female", "Feminino", inplace=True)
df.replace("Male", "Masculino", inplace=True)

# Raça
df.replace("White", "Branca", inplace=True)
df.replace("Black or African-American", "Preta", inplace=True)
df.replace("Asian or Pacific Islander", "Asiática", inplace=True)
df.replace("Mixed race", "Raça Mista", inplace=True)
df.replace("Native American/American Indian", "Indígena", inplace=True)
df.replace("Other", "Outro", inplace=True)
df.replace("Refused", "Não Informado", inplace = True)

# Estado Civil
df.replace("Married", "Casado", inplace=True)
df.replace("Divorced", "Divorciado", inplace=True)
df.replace("Widowed", "Viúvo", inplace=True)
df.replace("Never been married", "Solteiro", inplace=True)
df.replace("Separated", "Solteiro", inplace=True)
df.replace("Single", "Solteiro", inplace=True)
df.replace("Living with a partner", "Outros", inplace=True)
df.replace("Don't know", "Não Informado", inplace=True)

# Educação

df.replace("Some college, no 4-year degree", "Ensino médio completo", inplace=True)
df.replace("High school incomplete", "Ensino médio incompleto", inplace=True)
df.replace("High school graduate", "Ensino médio completo", inplace=True)
df.replace("College graduate", "Ensino superior completo", inplace=True)
df.replace("Post-graduate training/professional school after college", "Pós-graduação", inplace=True)
df.replace("Technical, trade or vocational school AFTER high school", "Ensino técnico", inplace=True)

# Emprego

df.replace("Employed full-time", "Empregado (Integral)", inplace=True)
df.replace("Employed part-time", "Empregado (Meio Período)", inplace=True)
df.replace("Retired", "Aposentado", inplace=True)
df.replace("Have own business/self-employed", "Autônomo", inplace = True)
df.replace("Student", "Estudante", inplace=True)
df.replace("Disabled", "Desabilitado", inplace=True)
df.replace("Not employed for pay", "Desempregado", inplace = True)

# Salario

df.replace("Less than $10,000", "Menos que $10,000", inplace=True)
df.replace("$20,000 to under $30,000", "$10,000 a $20,000", inplace=True)
df.replace("$40,000 to under $50,000", "$40,000 a $50,000", inplace=True)
df.replace("$10,000 to under $20,000", "$10,000 a $20,000", inplace=True)
df.replace("$50,000 to under $75,000", "$50,000 a $75,000", inplace=True)
df.replace("$100,000 to under $150,000", "$100,000 a $150,000", inplace=True)
df.replace("$75,000 to under $100,000", "$75,000 a $100,000", inplace=True)
df.replace("$30,000 to under $40,000", "$30,000 a $40,000", inplace=True)

# Último livro lido

df.replace("Purchased the book", "Comprou o livro", inplace=True)
df.replace("Borrowed the book from a friend or family member", "Pegou emprestado de alguém", inplace=True)
df.replace("Borrowed the book from a library", "Pegou numa biblioteca", inplace=True)
df.replace("Got the book some other way", "Pegou de outra forma", inplace=True)
df.replace("8", "Pegou de outra forma", inplace=True)
df.replace("9", "Pegou de outra forma", inplace=True)
```

# **Tratando valores nulos**

```python
df['Educação'].fillna("Não Informado", inplace=True)
df['Leu algum livro físico nos últimos 12 meses?'].fillna("Não Informado", inplace=True)
df['Leu algum audiolivro nos últimos 12 meses?'].fillna("Não Informado", inplace=True)
df['Leu algum e-book nos últimos 12 meses?'].fillna("Não Informado", inplace=True)
df['No último livro que você leu, você…'].fillna("Pegou de outra forma", inplace=True)
```
